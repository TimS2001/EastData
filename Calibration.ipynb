```python

import numpy as np
import matplotlib.pyplot as plt
#

from CalibrationFunc import GetData, plotAndFind, plotWithout, GetKoefs
```

Work with Calibration file.
Data is used from Read_scale_data directory

There is Energy parameter, that have to same as a energy from files

```python
Energies = []
Channels = []
```
